### Loading videos and splitting into frames 

```python
import cv2
import os

video_dir = os.path.join(os.getcwd(), "<videos_dir>")

vid_files = os.listdir(video_dir)
output_dir = "frames"
for vid in vid_files:
    # if the filename ends with np4 or mov, then process the file
    if(not vid.endswith(".mp4") and not vid.endswith(".mov")):
        continue
    if not os.path.exists(output_dir):
        os.mkdir(output_dir)
    frames_folder = os.path.join(output_dir, f"__{vid.split('.')[0]}__")
    frame_folder_path = os.path.join(os.getcwd(), frames_folder)
    if(not os.path.exists(frame_folder_path)):
        os.mkdir(frame_folder_path)
    
    # process the video using openCV
    cap = cv2.VideoCapture(os.path.join(video_dir, vid))
    fps = cap.get(cv2.CAP_PROP_FPS)
    print(f"Video {vid} FPS:", fps)
    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        # write each frame into a separate jepg image and save it
        cv2.imwrite(f"{frame_folder_path}/__{vid.split('.')[0]}__frame_{frame_count:06d}.jpg", frame)
        frame_count += 1
    cap.release()
    print(f"Saved {frame_count} frames. to {frame_folder_path}")
```
